```python
import pandas as pd
df = pd.read_csv("home_prices.csv")
df.head()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(df[['area_sqr_ft']],df['price_lakhs'])
```

```python
model.predict([[1500]])
```

```python
model.predict([[1350],[700]])
```

```python
#slope(m),  intercept(b)
model.coef_,model.intercept_
```

```python
m=model.coef_[0]
m
```

```python
b = model.intercept_
b
```

```python
#y=mx+b
m*1500+b
```
